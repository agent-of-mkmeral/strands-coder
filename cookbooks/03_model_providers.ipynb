## 🎯 Why Different Providers?

Each model provider has strengths:

| Provider | Strengths | Best For |
|----------|-----------|----------|
| Amazon Bedrock | Enterprise-grade, multiple models | Production, AWS integration |
| OpenAI | GPT-4, widely used | General purpose, rapid prototyping |
| Anthropic | Claude models, safety-focused | Complex reasoning, safe AI |
| Ollama | Free, local, private | Learning, offline use, privacy |

Strands makes it easy to switch between them!

## 🔷 Provider 1: Amazon Bedrock (Default)

Bedrock is the default provider - it offers access to multiple foundation models.

```python
from strands import Agent
from strands.models import BedrockModel

# Configure Bedrock model explicitly
bedrock_model = BedrockModel(
    model_id="us.anthropic.claude-sonnet-4-20250514-v1:0",  # Claude Sonnet 4
    temperature=0.3,  # Lower = more focused/deterministic
    streaming=True,   # Enable streaming responses
)

# Create agent with Bedrock
agent = Agent(model=bedrock_model)

# Note: Requires AWS credentials configured
# Run 'aws configure' or set AWS_ACCESS_KEY_ID and AWS_SECRET_ACCESS_KEY

response = agent("Explain what cloud computing is in simple terms.")
print("\n" + "="*50)
print("✅ Bedrock response complete!")
```

### Bedrock Model Options

Common Bedrock models:
- `us.anthropic.claude-sonnet-4-20250514-v1:0` - Claude Sonnet 4 (balanced)
- `us.amazon.nova-pro-v1:0` - Amazon Nova Pro
- `us.amazon.nova-lite-v1:0` - Amazon Nova Lite (faster)

## 🟢 Provider 2: OpenAI

Use OpenAI's GPT models. Requires an API key.

```python
import os
from strands import Agent
from strands.models.openai import OpenAIModel

# Set your OpenAI API key
# os.environ["OPENAI_API_KEY"] = "your-api-key-here"  # Uncomment and add your key

# Check if API key is set
if os.environ.get("OPENAI_API_KEY"):
    openai_model = OpenAIModel(
        model_id="gpt-4o-mini",  # Cost-effective option
        params={"temperature": 0.7}
    )
    
    agent = Agent(model=openai_model)
    response = agent("What are the three laws of robotics?")
    print("\n" + "="*50)
    print("✅ OpenAI response complete!")
else:
    print("⚠️ Set OPENAI_API_KEY to use OpenAI")
    print("Example: os.environ['OPENAI_API_KEY'] = 'sk-...'")
```

## 🟣 Provider 3: Anthropic

Anthropic's Claude models are known for safety and reasoning.

```python
import os
from strands import Agent
from strands.models.anthropic import AnthropicModel

# Set your Anthropic API key
# os.environ["ANTHROPIC_API_KEY"] = "your-api-key-here"  # Uncomment and add your key

if os.environ.get("ANTHROPIC_API_KEY"):
    anthropic_model = AnthropicModel(
        model_id="claude-sonnet-4-20250514",
        params={"temperature": 0.5, "max_tokens": 1024}
    )
    
    agent = Agent(model=anthropic_model)
    response = agent("Explain the concept of recursion with a simple example.")
    print("\n" + "="*50)
    print("✅ Anthropic response complete!")
else:
    print("⚠️ Set ANTHROPIC_API_KEY to use Anthropic")
    print("Example: os.environ['ANTHROPIC_API_KEY'] = 'sk-ant-...'")
```

## 🦙 Provider 4: Ollama (Local Models)

Ollama runs models locally - **free and private!**

### Setup Ollama:
1. Install from https://ollama.ai
2. Run: `ollama pull llama3.2` (or another model)
3. Ollama runs automatically in the background

```python
from strands import Agent
from strands.models.ollama import OllamaModel

# Create Ollama model (requires Ollama running locally)
try:
    ollama_model = OllamaModel(
        host="http://localhost:11434",
        model_id="llama3.2"  # Use llama3 if llama3.2 not available
    )
    
    agent = Agent(model=ollama_model)
    response = agent("What is the Fibonacci sequence?")
    print("\n" + "="*50)
    print("✅ Ollama response complete!")
except Exception as e:
    print(f"⚠️ Ollama error: {e}")
    print("\nTo use Ollama:")
    print("1. Install: https://ollama.ai")
    print("2. Run: ollama pull llama3.2")
    print("3. Make sure ollama is running")
```

## 🔄 Switching Between Providers

Here's a helper function to easily switch providers:

```python
import os
from strands import Agent

def create_agent(provider: str = "bedrock", **kwargs):
    """Create an agent with the specified provider.
    
    Args:
        provider: One of 'bedrock', 'openai', 'anthropic', 'ollama'
        **kwargs: Additional arguments passed to the model
    """
    if provider == "bedrock":
        from strands.models import BedrockModel
        model = BedrockModel(
            model_id=kwargs.get("model_id", "us.anthropic.claude-sonnet-4-20250514-v1:0"),
            temperature=kwargs.get("temperature", 0.5)
        )
    
    elif provider == "openai":
        from strands.models.openai import OpenAIModel
        model = OpenAIModel(
            model_id=kwargs.get("model_id", "gpt-4o-mini"),
            params={"temperature": kwargs.get("temperature", 0.7)}
        )
    
    elif provider == "anthropic":
        from strands.models.anthropic import AnthropicModel
        model = AnthropicModel(
            model_id=kwargs.get("model_id", "claude-sonnet-4-20250514"),
            params={"temperature": kwargs.get("temperature", 0.5)}
        )
    
    elif provider == "ollama":
        from strands.models.ollama import OllamaModel
        model = OllamaModel(
            host=kwargs.get("host", "http://localhost:11434"),
            model_id=kwargs.get("model_id", "llama3.2")
        )
    
    else:
        raise ValueError(f"Unknown provider: {provider}")
    
    return Agent(model=model, **{k: v for k, v in kwargs.items() 
                                 if k not in ['model_id', 'temperature', 'host']})

# Example usage:
# agent = create_agent("openai", model_id="gpt-4o")
# agent = create_agent("ollama", model_id="mistral")

print("✅ Helper function ready!")
print("\nUsage: agent = create_agent('provider_name')")
print("Providers: 'bedrock', 'openai', 'anthropic', 'ollama'")
```

## ⚙️ Model Parameters

Common parameters you can tune:

| Parameter | Description | Range |
|-----------|-------------|-------|
| `temperature` | Creativity/randomness | 0.0 - 1.0 |
| `max_tokens` | Maximum response length | 1 - varies |
| `top_p` | Nucleus sampling | 0.0 - 1.0 |

### Temperature Guide:
- **0.0-0.3**: Focused, deterministic (facts, code)
- **0.4-0.7**: Balanced (general tasks)
- **0.8-1.0**: Creative (stories, brainstorming)

```python
from strands import Agent
from strands.models import BedrockModel

# Low temperature - focused and deterministic
focused_model = BedrockModel(
    model_id="us.anthropic.claude-sonnet-4-20250514-v1:0",
    temperature=0.1
)

# High temperature - creative
creative_model = BedrockModel(
    model_id="us.anthropic.claude-sonnet-4-20250514-v1:0",
    temperature=0.9
)

focused_agent = Agent(model=focused_model)
creative_agent = Agent(model=creative_model)

prompt = "Give me a name for a pet hamster."

print("🎯 Focused Agent (temp=0.1):")
response1 = focused_agent(prompt)

print("\n\n🎨 Creative Agent (temp=0.9):")
response2 = creative_agent(prompt)

print("\n" + "="*50)
print("Notice how the creative agent might give more unique names!")
```
