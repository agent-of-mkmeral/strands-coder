## 🤖 Step 3: Your First Agent

Let's create the simplest possible agent - just a few lines of code!

````python
from strands import Agent

# Create a basic agent with default settings
agent = Agent()

# Ask the agent a question
response = agent("What is Python and why is it popular for AI development?")

print("\n" + "="*50)
print("Agent Response Complete!")
````

## 🎯 Step 4: Understanding the Agent

What just happened? Let's break it down:

1. **`Agent()`** - Creates an agent instance with a default model
2. **`agent("...")`** - Sends your message to the AI model
3. The agent processes your request and streams the response

The Strands agent loop works like this:

```
Input → Reasoning (LLM) → Tool Selection → Tool Execution → Response
         ↑                                      ↓
         └──────────────────────────────────────┘
```

## 🛠️ Step 5: Adding Tools

Agents become powerful when you give them tools! Let's add a calculator tool.

````python
from strands import Agent
from strands_tools import calculator

# Create an agent with the calculator tool
agent_with_tools = Agent(tools=[calculator])

# Now ask a math question - the agent will use the calculator!
response = agent_with_tools("What is the square root of 1764?")

print("\n" + "="*50)
print("✅ The agent used the calculator tool to solve this!")
````

## ⏰ Step 6: More Built-in Tools

Strands comes with many useful tools. Let's try the `current_time` tool.

````python
from strands import Agent
from strands_tools import calculator, current_time

# Create an agent with multiple tools
agent = Agent(tools=[calculator, current_time])

# Ask about time and math in the same request!
response = agent("""
I have two questions:
1. What is the current time?
2. If I have 42 apples and want to divide them among 6 friends, how many does each person get?
""")

print("\n" + "="*50)
print("✅ The agent used multiple tools to answer both questions!")
````

## 📊 Step 7: Understanding Agent Results

The agent returns an `AgentResult` object with useful information.

````python
from strands import Agent
from strands_tools import calculator

agent = Agent(tools=[calculator])
result = agent("What is 15 factorial?")

# Access the metrics
print("\n" + "="*50)
print("📊 Agent Metrics:")
print(f"   Total duration: {result.metrics.accumulated_metrics.get('latencyMs', 0)/1000:.2f} seconds")
print(f"   Input tokens: {result.metrics.accumulated_usage.get('inputTokens', 0)}")
print(f"   Output tokens: {result.metrics.accumulated_usage.get('outputTokens', 0)}")
````

## 🎓 Step 8: System Prompts

You can customize your agent's personality using system prompts!

````python
from strands import Agent

# Create a pirate agent! 🏴‍☠️
pirate_agent = Agent(
    system_prompt="""You are a friendly pirate assistant named Captain Code.
    Always speak like a pirate (arr, matey, landlubber, etc.) but still be helpful and accurate.
    Keep responses concise."""
)

response = pirate_agent("Explain what a variable is in programming.")

print("\n" + "="*50)
print("🏴‍☠️ Captain Code explained programming concepts!")
````
