## 🤖 Step 3: Your First Agent

Now let's create and use an agent! An agent is an AI that can:
- Have conversations
- Use tools to take actions
- Maintain context across messages

````python
from strands import Agent

# Create a simple agent
agent = Agent(
    system_prompt="You are a helpful assistant. Be concise and friendly."
)

# Chat with the agent!
response = agent("Hello! What can you help me with today?")

print("\n✅ Agent responded!")
````

### 💬 Multi-turn Conversations

Agents remember context across multiple messages:

````python
from strands import Agent

agent = Agent(
    system_prompt="You are a math tutor. Help students understand concepts step by step."
)

# First message
agent("I'm learning about fractions. What are they?")

# Follow-up (agent remembers context!)
agent("Can you give me a simple example?")

# Another follow-up
agent("How do I add two fractions together?")

print(f"\n📊 Conversation has {len(agent.messages)} messages")
````

---
## 🔧 Step 4: Using Tools

Tools give agents special abilities! Let's use the built-in calculator tool.

````python
from strands import Agent
from strands_tools import calculator

# Create an agent with the calculator tool
agent = Agent(
    tools=[calculator],
    system_prompt="You are a helpful math assistant."
)

# Ask a math question - the agent will use the calculator!
response = agent("What is 1234 * 5678?")

print("\n✅ Agent used the calculator tool!")
````

### 🔧 Multiple Tools

Agents can use multiple tools together:

````python
from strands import Agent
from strands_tools import calculator, current_time

# Agent with multiple tools
agent = Agent(
    tools=[calculator, current_time],
    system_prompt="You are a helpful assistant with access to a calculator and clock."
)

# The agent decides which tool(s) to use
response = agent("What time is it, and how many minutes until 5 PM?")

print("\n✅ Agent can use multiple tools!")
````

---
## 🧠 Step 5: Understanding the Agent Loop

When you send a message to an agent, here's what happens:

```
┌─────────────────────────────────────────────────────────────┐
│                      AGENT LOOP                              │
├─────────────────────────────────────────────────────────────┤
│  1. User sends message                                       │
│           ↓                                                  │
│  2. Agent sends to LLM with tools                           │
│           ↓                                                  │
│  3. LLM decides: respond OR use tool                        │
│           ↓                                                  │
│  4a. If tool: Execute → Send result back → Go to step 3     │
│  4b. If response: Return to user                            │
└─────────────────────────────────────────────────────────────┘
```

The agent keeps looping until it has a final response!

## 🎯 Step 6: Streaming Responses

For long responses, you can stream the output in real-time:

````python
from strands import Agent

agent = Agent(system_prompt="You are a creative storyteller.")

# Stream the response
print("Streaming response:\n")
for event in agent.stream("Tell me a very short story about a robot learning to cook."):
    if "data" in event:
        print(event["data"], end="", flush=True)

print("\n\n✅ Streaming complete!")
````
